```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
not_cust_h1 = [14.7,13,11.8,11.5,10.6,9.1,8.7,7.7,7.3,5.5]
cust_h1 = [2.8,5.3,7.3,7.7,9.1,11.4,12,13.5,14.1,16.8]
decile = np.arange(10,0,-1)
d = {'not_customers_half_1' : not_cust_h1, 'customers_half_1' : cust_h1, 'decile': decile }
df = pd.DataFrame(d)
df
```

```python
not_cust_h2 = [14.9,13.2,12.2,11.6,10.6,9.5,7.9,7.8,6.9,5.4]
cust_h2 = [2.8,5.3,6.7,7.5,9.2,10.8,13.1,13.3,14.6,16.8]
decile = np.arange(10,0,-1)
d2 = {'not_customers_half_2' : not_cust_h1, 'customers_half_2' : cust_h1, 'decile': decile }
df2 = pd.DataFrame(d2)
df2
```

```python
sns.set(style="whitegrid")
bar_width = 0.4
opacity = 0.8

# Plot for the first data frame (df)
fig1, ax1 = plt.subplots(figsize=(10, 6))

bar1 = ax1.bar(df['decile'] - bar_width/2, df['not_customers_half_1'], bar_width, label='Not Customers Half 1', alpha=opacity)
bar2 = ax1.bar(df['decile'] + bar_width/2, df['customers_half_1'], bar_width, label='Customers Half 1', alpha=opacity)

ax1.set_yticklabels(['{:,.0%}'.format(x/100) for x in ax1.get_yticks()])
ax1.set_xlabel('Decile')
ax1.set_ylabel('Percents')
ax1.set_title('Customer Scorecard Model Predictions: Training Group')
ax1.set_xticks(decile)
ax1.legend(loc='upper center')  # Adjust legend placement

for i, val in enumerate(df['not_customers_half_1']):
    plt.text(df['decile'][i] - bar_width/2, val + .25, f'{val:.1f}%', ha='center', va='bottom')

for i, val in enumerate(df['customers_half_1']):
    plt.text(df['decile'][i] + bar_width/2, val + .25, f'{val:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Plot for the second data frame (df2)
fig2, ax2 = plt.subplots(figsize=(10, 6))

bar1 = ax2.bar(df2['decile'] - bar_width/2, df2['not_customers_half_2'], bar_width, label='Not Customers Half 2', alpha=opacity)
bar2 = ax2.bar(df2['decile'] + bar_width/2, df2['customers_half_2'], bar_width, label='Customers Half 2', alpha=opacity)

ax2.set_xlabel('Decile')
ax2.set_ylabel('Percents')
ax2.set_title('Customer Scorecard Model Predictions: Test Group')
ax2.set_xticks(decile)
ax2.legend(loc='upper center')  # Adjust legend placement

for i, val in enumerate(df2['not_customers_half_2']):
    plt.text(df2['decile'][i] - bar_width/2, val + .25, f'{val:.1f}%', ha='center', va='bottom')

for i, val in enumerate(df2['customers_half_2']):
    plt.text(df2['decile'][i] + bar_width/2, val + .25, f'{val:.1f}%', ha='center', va='bottom')

plt.tight_layout()
plt.show()

```

```python
Variable = ["femaleness", "owns home", "income estimate", "length of residence", "type of dwelling","estimated home value", "fico score (2004)", "num cars in hh", "education levels (years)", "mdian hh income"]
SC = [-.161, .139, .075, -.113, -.089, .065, .049, .039, .033, .041]

d3 = {'variable' : Variable, 'Standardized Coef.' : SC }
df3 = pd.DataFrame(d3)
df3
```

```python
sns.set(style="whitegrid")

plt.figure(figsize=(10, 6))
bars = plt.barh(df3['variable'], df3['Standardized Coef.'], color=df3['Standardized Coef.'].apply(lambda x: 'green' if x > 0 else 'red'))

# Adding labels to the bars
for bar in bars:
    width = bar.get_width()
    plt.text(width - .01 if width > 0 else width + .01, bar.get_y() + bar.get_height()/2, f'{width:.3f}', ha='center', va='center', color='white' if width > 0 else 'black')

plt.xlabel('Standardized Coefficient')
plt.title('Positive and Negative Standardized Coefficients for Variables')
plt.grid(axis='x', linestyle='--', alpha=0.7)

plt.show()
```
